```python
# Valute in Euro
convert_GBP = 1.16 # Sterlina britannica
convert_DKK = 0.13 # Corona danese
convert_AUD = 0.61 # Dollaro australiano
convert_CHF = 1.04 # Franco svizzero
convert_USD = 0.91 # Dollaro americano
convert_PLN = 0.22 # Zloty polacco
convert_MXN = 0.05 # Peso messicano
convert_SEK = 0.09 # Corona svedese
convert_CAD = 0.67 # Dollaro canadese
convert_NOK = 0.09 # Corona norvegese
convert_JPY = 0.0065 # Yen giapponese
convert_CLP = 0.0011 # Peso cileno
convert_ARS = 0.003 # Peso argentino
```

```python
# Conversione Economic Impact in Euro
def converti_euro(value):
    if '£' in value:
        numeric_value = float(value.replace('£', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_GBP * 1e6)
    
    if '€' in value:
        numeric_value = float(value.replace('€', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * 1 * 1e6)
    
    if 'USD' in value:
        numeric_value = float(value.replace('USD', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_USD * 1e6)
    
    if 'MXN' in value:
        numeric_value = float(value.replace('MXN', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_MXN * 1e6)
    
    if 'AU$' in value:
        numeric_value = float(value.replace('AU$', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_AUD * 1e6)
    
    if 'ARS' in value:
        numeric_value = float(value.replace('ARS', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_ARS * 1e6)
    
    if 'CHF' in value:
        numeric_value = float(value.replace('CHF', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_CHF * 1e6)
    
    if 'DKK' in value:
        numeric_value = float(value.replace('DKK', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_DKK * 1e6)
    
    if 'SEK' in value:
        numeric_value = float(value.replace('SEK', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_SEK * 1e6)
    
    if 'CLP' in value:
        numeric_value = float(value.replace('CLP', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_CLP * 1e6)
    
    if 'JPY' in value:
        numeric_value = float(value.replace('JPY', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_JPY * 1e6)
    
    if 'PLN' in value:
        numeric_value = float(value.replace('PLN', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_PLN * 1e6)
    
    if 'NOK' in value:
        numeric_value = float(value.replace('NOK', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_NOK * 1e6)
    
    if 'CAD' in value:
        numeric_value = float(value.replace('CAD', '').replace(' million', '').replace(' ', ''))
        return '{:.0f}'.format(numeric_value * convert_CAD * 1e6)
    
    return value  # Se non corrisponde a nessuna delle valute, restituisce il valore originale
    
```

```python
# Update colonna Economic Impact
df['Economic_Impact'] = df['Economic_Impact'].apply(converti_euro)
```
